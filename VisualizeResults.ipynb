```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2
import os
```

# YOLOv8 Training Results

### Metrics

```python
plt.figure(figsize=(15, 15))

confusion_matrix=cv2.imread(r'yolov8x_metrics\confusion_matrix.png')
plt.imshow(confusion_matrix)
plt.show()

plt.figure(figsize=(15, 15))
results_graphs=cv2.imread(r'yolov8x_metrics\results.png')
plt.imshow(results_graphs)
plt.show()
```

### model predictions

```python
yolov8_model = YOLO(r'weights\yolov8x.pt')
imgs_path = r'original_data_sample'
output_path = r'original_data_sample\original_sample_output_8x'


for img in os.listdir(imgs_path):
    results = yolov8_model.predict(os.path.join(imgs_path, img), iou=0.45, conf=0.15,verbose=False)
    output = results[0].plot()
    original_img = results[0].orig_img
    compo = cv2.hconcat([original_img, output])
    cv2.imwrite(os.path.join(output_path, img), compo)
    plt.figure(figsize=(5,10))
    plt.imshow(compo)
    plt.show()
```

# YOLOv11x Single Class Training

## Metrics (Training in progress)

#### In this approach the classification and segmentation tasks will be separated.
#### The chromosomes image will go through segmantation model first to extract all chromosomes instatnces.
#### Then each instance will go through classification model to determine the class of this chromosome.

### Initial Results

```python
yolov11_model_single = YOLO(r'weights\yolov11x_single_class.pt')
imgs_path = r'original_data_sample'
output_path = r'original_data_sample\original_sample_output_11x_single_cls'


for img in os.listdir(imgs_path):
    results = yolov11_model_single.predict(os.path.join(imgs_path, img), iou=0.5,verbose=False)
    output = results[0].plot()
    original_img = results[0].orig_img
    compo = cv2.hconcat([original_img, output])
    cv2.imwrite(os.path.join(output_path, img), compo)
    plt.figure(figsize=(5,10))
    plt.imshow(compo)
    plt.show()
    plt.close()
```
